Data Visualisation - How to Plot a Scatter Bubble Chart by Plotly.
Unlike a scatterplot that can only show the relationship between 2 variables, a Scatter Bubble Chart (SBC) can depict the relationships of three or four variables. For example, the SBC below shows the relationship between Life Expectancy and GDP per capita in y and x axis, but using color to differentiate between continents and using size to differentiate between population sizes. Thus, it does not only show a linear relationship between life expectancy and GDP per capita (in log scale), but it can also show a lower (higher) GDP per capita in Africa (North America) and a larger population size in two Asian countries. This SBC can hardly be produced by Excel, but it can be produced in just a few lines of codes by Plotly as shown below:
Scatter Bubble Chart Life Expectancy v GDP per capita 2015.png

```python
! pip install plotly==5.0.0
import plotly.express as px
```

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive/')
data = pd.read_csv("drive/MyDrive/Colab Notebooks/life-expectancy-vs-gdp-per-capita.csv")
data.head(5)
```

```python
from plotly.figure_factory import create_table
datatable = create_table(data.head(10))
datatable.show()
```

```python
df2015 = data.query("Year==2015")
fig2015 = px.scatter(df2015, x="GDP per capita", y="Life expectancy")
fig2015.show()
```

```python
df2015 = data.query("Year==2015")
fig2015 = px.scatter(df2015, x="GDP per capita", y="Life expectancy", log_x=True)
fig2015.show()
```

```python
df2015 = data.query("Year==2015")
fig2015 = px.scatter(df2015, x="GDP per capita", y="Life expectancy", log_x=True, size="Total population", size_max=60, color="Continent", hover_name="Country")
fig2015.show()
```
